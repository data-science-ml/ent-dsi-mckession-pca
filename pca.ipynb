```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_boston
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
iris = load_iris()
```

```python
iris.target[:5]
```

```python
iris.data[:5]
```

```python
km = KMeans(n_clusters=3).fit(iris.data, iris.target)
```

```python
iris_2d = PCA(n_components=2).fit_transform(iris.data)
```

```python
plt.scatter(iris_2d[:,0], iris_2d[:,1], c=km.labels_)
plt.show()
```

```python
plt.scatter(iris_2d[:,0], iris_2d[:,1], c=iris.target)
plt.show()
```

```python
boston = load_boston()
```

```python
X = boston.data
y = boston.target
```

```python
# 13 cols
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression().fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
for dim in range(2, 14):
    pca = PCA(n_components=dim).fit(X_train)
    xtr = pca.transform(X_train)
    xts = pca.transform(X_test)
    model1 = LinearRegression().fit(xtr, y_train)
    print(dim, model1.score(xts, y_test))
```
